```python
import os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as mplcm
import matplotlib.colors as colors

font = {'size'   : 18}
matplotlib.rc('font', **font)
```

```python
RESULT_PATH = '../sh_results/CASP12'

RANKS_METRICS = [
    'z_score',
    'rank',
]

OTHER_METRICS = [
    'mse',
    'r2',
    'pearson',
    'spearman'
]

APPROACHES = [
    'global',
    'decoy'
]

SCORE_TYPES = [
    'global',
    'local'
]

ALGORITHMS = [
    's5_reg_drp_casp_8_11_25_v4.eval',
    's5scoring_reg_drp_casp_8_11_25_v1.eval',
    's5w_reg_drp_casp_8_11_25_v4.eval',
#     's5_d0.05_reg0.01_o9_v1.eval',
#     's5_d0.2_reg0.01_o49_v1.eval'
]

NAMES = [
    'S-GCN(5)',
    'S-GCN'+ r'$_{\mathrm{s}}$' +'(5)',
    'S-GCN'+ r'$_{\mathrm{r}}$' +'(5)',
    'S-GCN(3)',
    'S-GCN(7)'
]

COLORS = ['red', 'blue', 'green']

SIGNS = ['-', '-', '-']

RANKS_METRICS_NAMES = ['z-score', 'rank']
OTHER_METRICS_NAMES = ['MSE', r'$R^2$', 'Pearson\'s ' + r'$r$', 'Sparman\'s ' + r'$\rho$']
APPROACHES_NAMES = ['global', 'per-residue']
SCORE_TYPES_NAMES = ['global', 'local']


dfs = {
    alg: pd.read_csv(os.path.join(RESULT_PATH, alg))
    for alg in ALGORITHMS
}
```

```python
def compute_means(all_dfs, alg, metric):
    dfs = [v for k, v in all_dfs.items()]
    min_len = np.min([len(df) for df in dfs])
    trajectories = np.stack([df[metric].values[:min_len] for df in dfs])
    means = trajectories
    return means

def plot_rank_metrics(dfs, path=None):
    fig1, axes = plt.subplots(1, 2, figsize=(20, 10))
    for i, metric in enumerate(RANKS_METRICS):
        for j, alg in enumerate(ALGORITHMS):
            m = dfs[alg][metric].values
            axes[i].plot(m, label=NAMES[j], color=COLORS[j], linestyle=SIGNS[j], linewidth=2)
        axes[i].set_title(RANKS_METRICS_NAMES[i])
        axes[i].legend()
        axes[i].grid(alpha=0.7)
    plt.tight_layout()
    if path:
        plt.savefig(path)
    plt.show()
    
def plot_other_metrics(dfs, path=None):
    fig, axes = plt.subplots(4, 4, figsize=(20, 20))
    for i, metric in enumerate(OTHER_METRICS):
        for j, score_type in enumerate(SCORE_TYPES):
            for k, approach in enumerate(APPROACHES):
                x1 = 2 * j + k
                x2 = i
                a = axes[x1][x2]
                full_metric_name = '{}_{}_{}'.format(metric, score_type, approach)
                title = '{} {},\n{} scores'.format(APPROACHES_NAMES[k], OTHER_METRICS_NAMES[i], SCORE_TYPES_NAMES[j])
                for n, alg in enumerate(ALGORITHMS):
                    m = dfs[alg][full_metric_name].values
                    a.plot(m, linestyle=SIGNS[n], color=COLORS[n], label=NAMES[n], linewidth=2)
                a.set_title(title)
                a.legend()
                a.grid(alpha=0.7)
    plt.tight_layout()
    if path:
        plt.savefig(path)
    plt.show()
```

```python
# plot_rank_metrics(dfs, path='../sh_latex/img_TMP/SI_5_ranks.pdf')
plot_other_metrics(dfs, path='../sh_latex/img_TMP/SI_5_others.pdf')
```

```python
TABLE_METRICS_NAMES = [
    'z_score', 
    'rank', 
    'mse_global_global',
    'mse_global_decoy',
    'mse_local_global',
    'mse_local_decoy',
    'r2_global_global',
    'r2_global_decoy',
    'r2_local_global',
    'r2_local_decoy',
    'pearson_global_global',
    'pearson_global_decoy',
    'pearson_local_global',
    'pearson_local_decoy',
    'spearman_global_global',
    'spearman_global_decoy',
    'spearman_local_global',
    'spearman_local_decoy',
]

KEY_METRIC = 'pearson_local_global'
```

```python
def find_best(all_dfs, alg, metric, chooser=np.argmax):
    vals = all_dfs[alg][metric].values
    return chooser(vals)
```

```python
best2output = []
for alg in ALGORITHMS:
    best_epoch = find_best(dfs, alg, KEY_METRIC, chooser=np.argmax)
    results = []
    for mname in TABLE_METRICS_NAMES:
        res = dfs[alg][mname].values[best_epoch]
        results.append(res)
    result = {
        'alg': alg,
        'best': best_epoch,
        'metric': results[12],
        'output': ','.join(map('{:.4f}'.format, results))
    }
    best2output.append(result)
```
